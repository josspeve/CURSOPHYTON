## 09_NonParametric_Mann_whitney and Exercise

```python
def import_csv(file):
    import pandas as pd
    return pd.read_csv(file)
```

```python
path="Terminos_lagoon_TA_DIC_2023_RawData.csv"

data=import_csv(path)
```

MAnn-Whitney U test

```python
from scipy.stats import mannwhitneyu    

def mann_whitney_u_test(data1, data2, alpha=0.05):
    result = mannwhitneyu(data1, data2, alternative='two-sided')
    
    print(f"U statistic: {result.statistic}") 
    print(f"P-value: {result.pvalue}")
    
    if result.pvalue < alpha:
        print("The difference is statistically significant.")
    else:
        print("The difference is not statistically significant.")
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

# Create a boxplot
sns.boxplot(x="season", y='ta_micromol_kg', data=data, palette="Set3")

# Add title and labels
plt.title('Boxplot of TA by Season')
plt.xlabel('Season')
plt.ylabel('TA (μmol/kg)')

# Show the plot
plt.show()
```

```python
ta_dry = data.loc[data["season"] == "Dry", "ta_micromol_kg"]
ta_rainy = data.loc[data["season"] == "Rainy", "ta_micromol_kg"]

# Perform the Mann-Whitney U test for TA between dry and rainy seasons
mann_whitney_u_test(ta_dry, ta_rainy)
```

```python
# Filter the salinity data by season
sal_dry = data.loc[data["season"] == "Dry", "sal_psu"]
sal_rainy = data.loc[data["season"] == "Rainy", "sal_psu"]

# Perform the Mann-Whitney U test for salinity between dry and rainy seasons
mann_whitney_u_test(sal_dry, sal_rainy)
```
