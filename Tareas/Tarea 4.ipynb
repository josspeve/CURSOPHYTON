## 06_plot_descriptive_statistics_matplotib

Plot histogram of DIC and TA


Importar archivo CSV como DataFrame

```python
def import_csv_to_dataframe(file_path):
    import pandas as pd
    return pd.read_csv(file_path)
```

```python
# Path to the file to be imported   

path= "Terminos_lagoon_TA_DIC_2023_RawData.csv"

# Import the CSV file as a DataFrame
data = import_csv_to_dataframe(path)
```

## Histogram of DIC and TA

```python
import matplotlib.pyplot as plt 
# Seleccionar data

sal_data = data["sal_psu"]

plt.hist(sal_data, bins=10, color='pink', edgecolor= "black")
plt.xlabel("Salinity (PSU)")
plt.ylabel("Probablity")             
plt.show()  
```

## Normalizado

```python
data_norm = data["sal_psu"]

plt.hist(data_norm, bins=10, density=True, color='purple', edgecolor= "black")
plt.xlabel("Salinity (PSU)")
plt.ylabel("Probablity")             
plt.show()
```

# Boxplot 

```python
do_data= data["do_mg_l"]    

plt.boxplot(do_data)
plt.xlabel("DO  (mg/L)")
plt.ylabel("Concentration")
plt.title("Boxplot of DO Concentration")
plt.show()
```

## Boxplot Salinity 

```python
data_b = data["sal_psu"]

fig2, ax2 = plt.subplots()

ax2.set_title('Notched boxes')
ax2.boxplot(data_b, notch=True)         
plt.ylabel("Salinity ($P$SU)", fontsize=14)
plt.show()
```

# EJERCICIO 

```python
plt.hist(data["ta_micromol_kg"].dropna(), bins=30, alpha=0.6, label="TA (mol/kg)", color='blue')
plt.hist(data["dic_micromol_kg"].dropna(), bins=30, alpha=0.6,  label="DIC (mol/kg)", color='orange')
plt.xlabel("Concentration (mol/kg)")   
plt.ylabel("Frequency")
plt.title("Histogram of TA and DIC")        
plt.legend()
plt.show()
```
